```python
import numpy as np
import matplotlib.pyplot as plt
import math
import sys

def calculate_math_operation():
    print("Welcome to the Mathematical Calculator, comrades!")
    while True:
        print("Please choose a mathematical operation:")
        print("1. Perform arithmetic operations (addition, subtraction, multiplication, division)")
        print("2. Perform geometric operations (rectangle, square, triangle, circle, parallelogram)")
        print("3. Perform graph plotting")
        print("4. Exit")

        choice = input("Enter your choice (1/2/3/4):")

        if choice == '1':
            perform_arithmetic_operation()
        elif choice == '2':
            perform_geometric_operation()
        elif choice == '3':
            perform_graph_plot()
        elif choice == '4':
            print("Thank you for using, Long live communism!")
            sys.exit()
        else:
            print("Invalid choice, please try again, comrades.")

def perform_arithmetic_operation():
    x = float(input("Enter the first number:"))
    y = float(input("Enter the second number:"))
    operation = input("Enter the operation (add, subtract, multiply, divide):")

    if operation in ["add", "subtract", "multiply", "divide"]:
        if operation == "add":
            print("Result:", x + y)
        elif operation == "subtract":
            print("Result:", x - y)
        elif operation == "multiply":
            print("Result:", x * y)
        elif operation == "divide":
            if y != 0:
                print("Result:", x / y)
            else:
                print("Error: Division by zero!")
    else:
        print("Invalid operation, please try again.")

def perform_geometric_operation():
    print("Welcome to the Geometric Calculator, striving for the cause of communism!")
    while True:
        print("Please choose a shape:")
        print("1. Rectangle")
        print("2. Square")
        print("3. Triangle")
        print("4. Circle")
        print("5. Parallelogram")
        print("6. Return to Main Menu")
        print("Type 'exit' to return to Main Menu.")

        shape_choice = input("Enter your choice (1/2/3/4/5/6):")

        if shape_choice == '1':
            calculate_rectangle_area()
        elif shape_choice == '2':
            calculate_square_area()
        elif shape_choice == '3':
            calculate_triangle_area()
        elif shape_choice == '4':
            calculate_circle_area()
        elif shape_choice == '5':
            calculate_parallelogram_area()
        elif shape_choice == '6':
            break
        elif shape_choice.lower() == 'exit':
            return
        else:
            print("Invalid choice, please try again, comrades.")

#https://www.geeksforgeeks.org/python-program-to-find-area-of-rectangle/
def calculate_rectangle_area():
    length = validate_positive_float_input("Enter the length of the rectangle:")
    width = validate_positive_float_input("Enter the width of the rectangle:")
    area = length * width
    print("Area of the rectangle:", area)

#https://www.geeksforgeeks.org/calculating-areas-of-different-shapes-using-python/
def calculate_square_area():
    side = validate_positive_float_input("Enter the side length of the square:")
    area = side * side
    print("Area of the square:", area)

#https://www.geeksforgeeks.org/draw-circle-in-python-using-turtle/
def calculate_triangle_area():
    base = validate_positive_float_input("Enter the base length of the triangle:")
    height = validate_positive_float_input("Enter the height of the triangle:")
    area = 0.5 * base * height
    print("Area of the triangle:", area)

def calculate_circle_area():
    radius = validate_positive_float_input("Enter the radius of the circle:")
    area = math.pi * radius * radius
    print("Area of the circle:", area)

def calculate_parallelogram_area():
    base = validate_positive_float_input("Enter the base length of the parallelogram:")
    height = validate_positive_float_input("Enter the height of the parallelogram:")
    area = base * height
    print("Area of the parallelogram:", area)

def validate_positive_float_input(prompt):
    while True:
        try:
            value = float(input(prompt))
            if value <= 0:
                print("Value must be positive, please try again.")
            else:
                return value
        except ValueError:
            print("Invalid input, please enter a valid number.")

def perform_graph_plot():
    print("Plotting options:")
    print("1. Polynomial")
    print("2. Trigonometric")
    print("3. Linear and Quadratic Intersection")
    choice = input("Enter your choice (1/2/3):")

    if choice == '1':
        plot_polynomial_graph()
    elif choice == '2':
        plot_trigonometric_graph()
    elif choice == '3':
        plot_linear_quadratic_intersection()
    else:
        print("Invalid choice, comrades.")

#https://stackoverflow.com/questions/37352098/plotting-a-polynomial-using-matplotlib-and-coeffiecients
def plot_polynomial_graph():
    polynomial_str = input("Enter the polynomial equation (e.g., 3*x**2 + 4*x + 6):")
    x = np.linspace(-10, 10, 1000)
    try:
        y = eval(polynomial_str)
        plt.plot(x, y)
        plt.xlabel('x')
        plt.ylabel('y')
        plt.title('Graph: ' + polynomial_str)
        plt.grid(True)
        plt.show()
    except Exception as e:
        print("Error:", e)

#https://www.geeksforgeeks.org/plotting-sine-and-cosine-graph-using-matloplib-in-python/
def plot_trigonometric_graph():
    print("Please choose a trigonometric function:")
    print("1. sin(x)")
    print("2. cos(x)")
    print("3. tan(x)")
    trig_function_choice = input("Enter your choice (1/2/3):")

    if trig_function_choice in ['1', '2', '3']:
        x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
        if trig_function_choice == '1':
            y = np.sin(x)
            plt.title('Graph of sin(x)')
        elif trig_function_choice == '2':
            y = np.cos(x)
            plt.title('Graph of cos(x)')
        else:
            y = np.tan(x)
            y[:-1][np.diff(y) < 0] = np.nan
            plt.ylim(-10, 10)
            plt.title('Graph of tan(x)')
        plt.plot(x, y)
        plt.xlabel('x')
        plt.ylabel(trig_function_choice + '(x)')
        plt.grid(True)
        plt.show()
    else:
        print("Invalid choice, please try again, comrades.")

#https://stackoverflow.com/questions/28766692/how-to-find-the-intersection-of-two-graphs
#https://stackoverflow.com/questions/60832850/finding-the-intersect-between-a-quadratic-and-line
def plot_linear_quadratic_intersection():
    m_linear = float(input("Enter the slope (m) of the linear function:"))
    c_linear = float(input("Enter the intercept (c) of the linear function:"))

    a_quadratic = float(input("Enter the coefficient (a) of the quadratic function:"))
    b_quadratic = float(input("Enter the coefficient (b) of the quadratic function:"))
    c_quadratic = float(input("Enter the constant term (c) of the quadratic function:"))

    x = np.linspace(-10, 10, 1000)

    linear_y = linear_function(x, m_linear, c_linear)
    quadratic_y = quadratic_function(x, a_quadratic, b_quadratic, c_quadratic)

    plt.plot(x, linear_y, label='Linear Function')
    plt.plot(x, quadratic_y, label='Quadratic Function')

    idx = np.argwhere(np.diff(np.sign(linear_y - quadratic_y))).flatten()
    plt.plot(x[idx], linear_y[idx], 'ro', label='Intersection Point')

    plt.legend()

    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Intersection of Linear and Quadratic Functions')
    plt.grid(True)
    plt.show()

def linear_function(x, m, c):
    return m * x + c

def quadratic_function(x, a, b, c):
    return a * x**2 + b * x + c

if __name__ == "__main__":
    calculate_math_operation()
```
