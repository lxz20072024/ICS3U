```python
import matplotlib.pyplot as plt
import numpy as np
import sys

def show_sin_graph():
    x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
    y = np.sin(x)

    plt.grid()

    plt.xlabel("x")
    plt.ylabel("$sin(x)$")

    plt.ylim(-1, 1)
    plt.xlim(-2 * np.pi, 2 * np.pi)

    radian_multiples = [-2, -3/2, -1, -1/2, 0, 1/2, 1, 3/2, 2]
    radians = [n * np.pi for n in radian_multiples]
    radian_labels = ['$-2\pi$', '$-3\pi/2$', '$\pi$', '$-\pi/2$', '0', '$\pi/2$', '$\pi$', '$3\pi/2$', '$2\pi$']

    plt.xticks(radians, radian_labels)

    plt.title("$y = sin(x)$", fontsize=14)
    plt.plot(x, y)
    plt.show()
    print("Thank you for using!")
    sys.exit()

def show_cos_graph():
    x = np.arange(0, 5*np.pi, 0.1)
    y = np.cos(x)

    plt.plot(x, y, color='green')
    plt.xlabel('x')
    plt.ylabel('$cos(x)$')
    plt.title('$y = cos(x)$')
    plt.show()
    print("Thank you for using!")
    sys.exit()

def show_tan_graph():
    x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
    y = np.tan(x)

    y[:-1][np.diff(y) < 0] = np.nan

    plt.grid()

    plt.xlabel("x")
    plt.ylabel("$tan(x)$")

    plt.ylim(-10,10)
    plt.xlim(-2 * np.pi, 2 * np.pi)

    radian_multiples = [-2, -3/2, -1, -1/2, 0, 1/2, 1, 3/2, 2]
    radians = [n * np.pi for n in radian_multiples]
    radian_labels = ['$-2\pi$', '$-3\pi/2$', '$\pi$', '$-\pi/2$', '0', '$\pi/2$', '$\pi$', '$3\pi/2$', '$2\pi$']

    plt.xticks(radians, radian_labels)

    plt.title("$y = tan(x)$", fontsize=14)
    plt.plot(x, y)
    plt.show()
    print("Thank you for using!")
    sys.exit()

def main():
    print("Welcome to Graph Calculator!")
    while True:
        print("Please select an operation:")
        print("1. Enter graph mode")
        print("2. Quit")
        choice = input("Enter your choice (1/2): ")
        if choice == '1':
            print("Please select the graph to plot:")
            print("a. sin(x) graph")
            print("b. cos(x) graph")
            print("c. tan(x) graph")
            sub_choice = input("Enter your choice (a/b/c): ")
            if sub_choice == 'a':
                show_sin_graph()
            elif sub_choice == 'b':
                show_cos_graph()
            elif sub_choice == 'c':
                show_tan_graph()
            else:
                print("Invalid choice, please try again.")
        elif choice == '2':
            print("Thank you for using, goodbye!")
            sys.exit()
        else:
            print("Invalid choice, please try again.")

if __name__ == "__main__":
    main()
```
