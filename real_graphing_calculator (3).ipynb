```python
import numpy as np  # Importing numpy library for numerical operations
import matplotlib.pyplot as plt  # Importing matplotlib library for plotting graphs
import math  # Importing math library for mathematical functions
import sys  # Importing sys library for system-related operations
import random  # Importing random library for generating random numbers

# Function to handle the main mathematical operations
def calculate_math_operation():
    print("Welcome to the Mathematical Calculator, comrades!")  # Welcome message
    while True:
        # Displaying options for mathematical operations
        print("Please choose a mathematical operation:")
        print("1. Perform arithmetic operations (addition, subtraction, multiplication, division)")
        print("2. Perform geometric operations (rectangle, square, triangle, circle, parallelogram)")
        print("3. Perform graph plotting")
        print("4. Exit")

        choice = input("Enter your choice (1/2/3/4):")  # Taking user input for choice

        if choice == '1':
            perform_arithmetic_operation()  # Calling function for arithmetic operations
        elif choice == '2':
            perform_geometric_operation()  # Calling function for geometric operations
        elif choice == '3':
            perform_graph_plot()  # Calling function for graph plotting
        elif choice == '4':
            print("Thank you for using, Long live communism!")  # Exiting with gratitude
            sys.exit()  # Exiting the program
        else:
            print("Invalid choice, please try again, comrades.")  # Handling invalid input

# Function to perform arithmetic operations
def perform_arithmetic_operation():
    x = float(input("Enter the first number:"))  # Taking input for the first number
    y = float(input("Enter the second number:"))  # Taking input for the second number
    operation = input("Enter the operation (add, subtract, multiply, divide):")  # Taking input for the operation

    if operation in ["add", "subtract", "multiply", "divide"]:  # Checking if operation is valid
        if operation == "add":
            print("Result:", x + y)  # Performing addition
        elif operation == "subtract":
            print("Result:", x - y)  # Performing subtraction
        elif operation == "multiply":
            print("Result:", x * y)  # Performing multiplication
        elif operation == "divide":
            if y != 0:
                print("Result:", x / y)  # Performing division if denominator is not zero
            else:
                print("Error: Division by zero!")  # Handling division by zero
    else:
        print("Invalid operation, please try again.")  # Handling invalid operation

# Function to perform geometric operations
def perform_geometric_operation():
    print("Welcome to the Geometric Calculator, striving for the cause of communism!")  # Welcome message
    while True:
        # Displaying options for geometric shapes
        print("Please choose a shape:")
        print("1. Rectangle")
        print("2. Square")
        print("3. Triangle")
        print("4. Circle")
        print("5. Parallelogram")
        print("6. Return to Main Menu")
        print("Type 'exit' to return to Main Menu.")

        shape_choice = input("Enter your choice (1/2/3/4/5/6):")  # Taking user input for shape choice

        if shape_choice == '1':
            calculate_rectangle_area()  # Calculating area of rectangle
        elif shape_choice == '2':
            calculate_square_area()  # Calculating area of square
        elif shape_choice == '3':
            calculate_triangle_area()  # Calculating area of triangle
        elif shape_choice == '4':
            calculate_circle_area()  # Calculating area of circle
        elif shape_choice == '5':
            calculate_parallelogram_area()  # Calculating area of parallelogram
        elif shape_choice == '6':
            break  # Returning to main menu
        elif shape_choice.lower() == 'exit':
            return  # Exiting to main menu
        else:
            print("Invalid choice, please try again, comrades.")  # Handling invalid input

#https://www.geeksforgeeks.org/calculating-areas-of-different-shapes-using-python/ This one provided all the code I needed to calculate the area, and I copied part of it

# Function to calculate area of rectangle
def calculate_rectangle_area():
    length = validate_positive_float_input("Enter the length of the rectangle:")  # Taking input for length
    width = validate_positive_float_input("Enter the width of the rectangle:")  # Taking input for width
    area = length * width  # Calculating area
    print("Area of the rectangle:", area)  # Displaying area

# Function to calculate area of square
#https://www.geeksforgeeks.org/python-program-to-find-area-of-rectangle/ This is a rectangle-specific code, I copied some of the code as well as modified some of the
def calculate_square_area():
    side = validate_positive_float_input("Enter the side length of the square:")  # Taking input for side length
    area = side * side  # Calculating area
    print("Area of the square:", area)  # Displaying area

# Function to calculate area of triangle
def calculate_triangle_area():
    base = validate_positive_float_input("Enter the base length of the triangle:")  # Taking input for base length
    height = validate_positive_float_input("Enter the height of the triangle:")  # Taking input for height
    area = 0.5 * base * height  # Calculating area
    print("Area of the triangle:", area)  # Displaying area

# Function to calculate area of circle
#https://www.geeksforgeeks.org/draw-circle-in-python-using-turtle/ We have this circular code, I half copied the code as well as learned most of it and changed some of it as well
def calculate_circle_area():
    radius = validate_positive_float_input("Enter the radius of the circle:")  # Taking input for radius
    area = math.pi * radius * radius  # Calculating area
    print("Area of the circle:", area)  # Displaying area

# Function to calculate area of parallelogram
#https://www.w3resource.com/python-exercises/math/python-math-exercise-4.php Trapezoidal belongs to a very abstract graphics, so this code I copied as well as not yet to prevent problems I did not change, just some of the names unified a bit
def calculate_parallelogram_area():
    base = validate_positive_float_input("Enter the base length of the parallelogram:")  # Taking input for base length
    height = validate_positive_float_input("Enter the height of the parallelogram:")  # Taking input for height
    area = base * height  # Calculating area
    print("Area of the parallelogram:", area)  # Displaying area

# Function to validate positive float input
def validate_positive_float_input(prompt):
    while True:
        try:
            value = float(input(prompt))  # Converting input to float
            if value <= 0:
                print("Value must be positive, please try again.")  # Handling non-positive input
            else:
                return value  # Returning valid input
        except ValueError:
            print("Invalid input, please enter a valid number.")  # Handling invalid input

# Function to perform graph plotting
#https://stackoverflow.com/questions/37352098/plotting-a-polynomial-using-matplotlib-and-coeffiecients This one is the most basic Q&A, and I've borrowed from the overall format
def perform_graph_plot():
    print("Plotting options:")  # Displaying plotting options
    print("1. Polynomial")
    print("2. Trigonometric")
    print("3. Linear and Quadratic Intersection")
    choice = input("Enter your choice (1/2/3):")  # Taking user input for choice

    if choice == '1':
        plot_polynomial_graph()  # Plotting polynomial graph
    elif choice == '2':
        plot_trigonometric_graph()  # Plotting trigonometric graph
    elif choice == '3':
        plot_linear_quadratic_intersection()  # Plotting intersection of linear and quadratic functions
    else:
        print("Invalid choice, comrades.")  # Handling invalid input

# Function to plot polynomial graph
#https://www.geeksforgeeks.org/plotting-sine-and-cosine-graph-using-matloplib-in-python/ This one is a trigonometric graph drawing, drawing on the overall format
def plot_polynomial_graph():
    polynomial_str = input("Enter the polynomial equation (e.g., 3*x**2 + 4*x + 6):")  # Taking input for polynomial equation
    x = np.linspace(-10, 10, 1000)  # Generating x values
    try:
        y = eval(polynomial_str)  # Evaluating polynomial equation
        plt.plot(x, y)  # Plotting graph
        plt.xlabel('x')  # Labeling x-axis
        plt.ylabel('y')  # Labeling y-axis
        plt.title('Graph: ' + polynomial_str)  # Setting title
        plt.grid(True)  # Adding grid
        plt.show()  # Displaying graph
    except Exception as e:
        print("Error:", e)  # Handling errors

# Function to plot trigonometric graph
def plot_trigonometric_graph():
    print("Please choose a trigonometric function:")  # Displaying trigonometric function options
    print("1. sin(x)")
    print("2. cos(x)")
    print("3. tan(x)")
    trig_function_choice = input("Enter your choice (1/2/3):")  # Taking user input for choice

    if trig_function_choice in ['1', '2', '3']:
        x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)  # Generating x values
        if trig_function_choice == '1':
            y = np.sin(x)  # Calculating sine function
            plt.title('Graph of sin(x)')  # Setting title
        elif trig_function_choice == '2':
            y = np.cos(x)  # Calculating cosine function
            plt.title('Graph of cos(x)')  # Setting title
        else:
            y = np.tan(x)  # Calculating tangent function
            y[:-1][np.diff(y) < 0] = np.nan
            plt.ylim(-10, 10)
            plt.title('Graph of tan(x)')  # Setting title
        plt.plot(x, y)  # Plotting graph
        plt.xlabel('x')  # Labeling x-axis
        plt.ylabel(trig_function_choice + '(x)')  # Labeling y-axis
        plt.grid(True)  # Adding grid
        plt.show()  # Displaying graph
    else:
        print("Invalid choice, please try again, comrades.")  # Handling invalid input


# Function to plot intersection of linear and quadratic functions
#https://stackoverflow.com/questions/29904423/how-to-find-the-points-of-intersection-of-a-line-and-multiple-curves-in-python lit. turn to any doctor one can find when critically ill (idiom); fig. to try anyone or anything in a crisis
#https://stackoverflow.com/questions/28766692/how-to-find-the-intersection-of-two-graphs Copied some of the code but chopped 90% of it
#https://stackoverflow.com/questions/60832850/finding-the-intersect-between-a-quadratic-and-line Attempted to fix the code based on this at the time, and stopped when I realized I couldn't fix it anymore
def plot_linear_quadratic_intersection():
    m_linear = float(input("Enter the slope (m) of the linear function:"))  # Taking input for slope of linear function
    c_linear = float(input("Enter the intercept (c) of the linear function:"))  # Taking input for intercept of linear function

    a_quadratic = float(input("Enter the coefficient (a) of the quadratic function:"))  # Taking input for coefficient 'a' of quadratic function
    b_quadratic = float(input("Enter the coefficient (b) of the quadratic function:"))  # Taking input for coefficient 'b' of quadratic function
    c_quadratic = float(input("Enter the constant term (c) of the quadratic function:"))  # Taking input for constant term 'c' of quadratic function

    x = np.linspace(-10, 10, 1000)  # Generating x values

    linear_y = linear_function(x, m_linear, c_linear)  # Calculating y values for linear function
    quadratic_y = quadratic_function(x, a_quadratic, b_quadratic, c_quadratic)  # Calculating y values for quadratic function

    plt.plot(x, linear_y, label='Linear Function')  # Plotting linear function
    plt.plot(x, quadratic_y, label='Quadratic Function')  # Plotting quadratic function

    idx = np.argwhere(np.diff(np.sign(linear_y - quadratic_y))).flatten()  # Finding intersection points
    plt.plot(x[idx], linear_y[idx], 'ro', label='Intersection Point')  # Plotting intersection points

    plt.legend()  # Adding legend

    plt.xlabel('x')  # Labeling x-axis
    plt.ylabel('y')  # Labeling y-axis
    plt.title('Intersection of Linear and Quadratic Functions')  # Setting title
    plt.grid(True)  # Adding grid
    plt.show()  # Displaying graph

# Function for linear function
def linear_function(x, m, c):
    return m * x + c  # Linear function equation

# Function for quadratic function
def quadratic_function(x, a, b, c):
    return a * x**2 + b * x + c  # Quadratic function equation

# Main function
if __name__ == "__main__":
    calculate_math_operation()  # Calling main function
```
