```python
import matplotlib.pyplot as plt
import numpy as np
import sys

def show_sin_graph():
    x_min = float(input("Enter the minimum value of x-axis: "))
    x_max = float(input("Enter the maximum value of x-axis: "))
    num_points = int(input("Enter the number of points: "))

    x = np.linspace(x_min, x_max, num_points)
    y = np.sin(x)

    plt.grid()

    plt.xlabel("x")
    plt.ylabel("$sin(x)$")

    plt.ylim(-1, 1)
    plt.xlim(x_min, x_max)

    plt.title("$y = sin(x)$", fontsize=14)
    plt.plot(x, y)
    plt.show()
    print("Thank you for using! Exiting...")
    sys.exit()
# https://www.geeksforgeeks.org/plotting-sine-and-cosine-graph-using-matloplib-in-python/

def show_cos_graph():
    x_min = float(input("Enter the minimum value of x-axis: "))
    x_max = float(input("Enter the maximum value of x-axis: "))
    num_points = int(input("Enter the number of points: "))

    x = np.linspace(x_min, x_max, num_points)
    y = np.cos(x)

    plt.grid()

    plt.xlabel("x")
    plt.ylabel("$cos(x)$")

    plt.ylim(-1, 1)
    plt.xlim(x_min, x_max)

    plt.title("$y = cos(x)$", fontsize=14)
    plt.plot(x, y, color='green')
    plt.show()
    print("Thank you for using! Exiting...")
    sys.exit()
# https://stackoverflow.com/questions/61581552/how-can-i-graph-sine-and-cosine-functions-in-python

def show_tan_graph():
    x_min = float(input("Enter the minimum value of x-axis: "))
    x_max = float(input("Enter the maximum value of x-axis: "))
    num_points = int(input("Enter the number of points: "))

    x = np.linspace(x_min, x_max, num_points)
    y = np.tan(x)

    y[:-1][np.diff(y) < 0] = np.nan

    plt.grid()

    plt.xlabel("x")
    plt.ylabel("$tan(x)$")

    plt.ylim(-10,10)
    plt.xlim(x_min, x_max)

    plt.title("$y = tan(x)$", fontsize=14)
    plt.plot(x, y)
    plt.show()
    print("Thank you for using graphing calculator, goodbye")

# https://stackoverflow.com/questions/54505852/how-to-plot-tanx-with-pyplot-and-numpy

def main():
    print("Welcome to Graph Calculator!")
    while True:
        print("Please select an option:")
        print("1. Enter Graph Mode")
        print("2. Exit")
        choice = input("Enter your choice (1/2): ")
        if choice == '1':
            print("Please select the graph to plot:")
            print("a. Sin(x) Graph")
            print("b. Cos(x) Graph")
            print("c. Tan(x) Graph")
            sub_choice = input("Enter your choice (a/b/c): ")
            if sub_choice == 'a':
                show_sin_graph()
            elif sub_choice == 'b':
                show_cos_graph()
            elif sub_choice == 'c':
                show_tan_graph()
            else:
                print("Invalid choice, please select again.")
        elif choice == '2':
            print("Thank you for using! Goodbye!")
            sys.exit()
        else:
            print("Invalid choice, please select again.")

if __name__ == "__main__":
    main()
```
